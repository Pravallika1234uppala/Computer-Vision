```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,8))
    ax=fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
    plt.axis('off')
    plt.show()
    plt.close()
```

```python
internal_external_img = cv2.imread('Data/internal_external.png', 0)

display_img(internal_external_img)
```

## CONTOUR DETECTION:

```python
#useful for analysing shape and object detection and recognition.
#mode=
#RETR_CCOMP-way to find contour extarcts(internal and external)
#RETR_EXTERNAL - to grab only external contours
#method=
#CHAIN_APPROX_SIMPLE

contours,hierarchy = cv2.findContours(internal_external_img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
#hierarchy - bunch of lines that defines where contours are:

external_contours = np.zeros(internal_external_img.shape)
```

```python
display_img(external_contours)
```

```python
#shape of image - but all 0's (black)

#now to draw external contours

for i in range(len(contours)):
    #in range of no. of shapes
    if hierarchy[0][i][3]==-1:
    #point where contour is
    #last column defines internal or external (-1 is for external)
    #first element -0 (only 0 is prersent) 
    #second is no. of shapes(no. of contours)-i
        #external contour-drawing externals on external_contours
        cv2.drawContours(external_contours, contours, i, 255, -1)
        #filling white color with index
display_img(external_contours)
```

```python
internal_contours = np.zeros(internal_external_img.shape)

display_img(internal_contours)
```

```python
#shape of image - but all 0's (black)

#now to draw internal contours

for i in range(len(contours)):
    #in range of no. of shapes
    if hierarchy[0][i][3]!=-1:
    #point where contour is
    #last column defines internal or external (0 & 4 is for internal)
    #first element -0 (only 0 is prersent) 
    #second is no. of shapes(no. of contours)-i
        #internal contour-drawing internals on internal_contours
        cv2.drawContours(internal_contours, contours, i, 255, -1)
        #filling white color with index
display_img(internal_contours)
```
