### LAPLACIAN DERIVATIVE:

given by relationship

sec derivative f(regarding x)+sec derivative f(regarding y)

(these derivatives are nothing but sobel operators in x,y gradient)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
img = cv2.imread('Data/sudoku.jpg',0)
def display_img(img):
    fig=plt.figure(figsize=(12,8))
    ax=fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
    plt.axis('off')
    plt.show()
    plt.close()
```

```python
display_img(img)
```

```python
#applying laplacian:
laplacian_img=cv2.Laplacian(img, cv2.CV_64F)
display_img(laplacian_img)
```

```python
#can see x and y clearly(but has noise)

#so by blending sober x, y with Laplacian - to see image clearly

sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)

blended_img = cv2.addWeighted(sobel_x, 0.5, sobel_y, 0.5, 0)

display_img(blended_img)
```

```python
#applying thresholding and morphological operators:

kernel = np.ones((4,4), np.uint8)

gradient=cv2.morphologyEx(blended_img, cv2.MORPH_GRADIENT, kernel)

display_img(blended_img)
#gives us blurry image
```

```python
blendimg = cv2.addWeighted(blended_img, 0.3, laplacian_img, 0.7, 0)

display_img(blendimg)
```
