### GRADIENTS:

diff b/w erosion and dilation(edge)

can do many of these operations using morphological operators:

1. Object detection
2. Object Tracking
3. Image Classification

Image Gradient - Is directional change in intensity/color in an image.

Gradients can be calculated in specific direction:

> Normalized x direction gradient
> Normalized y direction gradient
> Normalized gradient Magnitude



sobel operator-uses 3*3 kernel 
3*3 kernel convolved with original image

```python
#Vertical edges

import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#displaying images:
img = cv2.imread('Data/sudoku.jpg',0)
def display_img(img):
    fig=plt.figure(figsize=(12,8))
    ax=fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
    plt.axis('off')
    plt.show()
    plt.close()
```

```python
display_img(img)
```

### (SOBEL-FELDMAN OPERATOR)NORMALIZED X-GRADIENT:

```python
#vertical edges
#depth-64f precision
#dx, dy-order of drrivative
#for normalized x gradient x-1 and for y-0


sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
```

```python
display_img(sobel_x)
```

### (SOBEL-FELDMAN OPERATOR)NORMALIZED Y-GRADIENT:

```python
#horizontal edges
#depth-64f precision
#dx, dy-order of drrivative
#for normalized x gradient x-1 and for y-0


sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
```

```python
display_img(sobel_y)
```

### (SOBEL-FELDMAN OPERATOR)NORMALIZED GRADIENT MAGNITUDE:

```python
# Computing Gradient Magnitude
gradient_mag = np.sqrt(sobel_x**2 + sobel_y**2)

display_img(gradient_mag)

# Normalizing the gray one to range [0, 255]-0, 255 value below are alpha beta values and normalized by min max way.
#gradient_mag = cv2.normalize(gradient_mag, None, 0, 255, cv2.NORM_MINMAX)

#display_img(gradient_mag)

# Converting to uint8 for display
#gradient_mag = np.uint8(gradient_mag)

#display_img(gradient_mag)
```
