```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def display_img(img):
    fig=plt.figure(figsize=(12,8))
    ax=fig.add_subplot(111)
    ax.imshow(img)
    plt.axis('off')
    plt.show()
    plt.close()
```

```python
flat_chess_bgr = cv2.imread('Data/flat_chessboard.png')

display_img(flat_chess_bgr)
```

## GRID DETECTION:

```python
#Finding corners to detect object first
#to find grid directly by builtin function in opencv
#here pattern is 7*7 as last one is not detected

found, corners = cv2.findChessboardCorners(flat_chess_bgr, (7,7))

if found:
    print(corners)
else:
    pass
```

```python
corners.shape
```

```python
#here we have 49 corner points x,y

#now drawing corners:(by feeding them)

flat_chess_bgr_cpy = flat_chess_bgr.copy()

cv2.drawChessboardCorners(flat_chess_bgr_cpy, (7,7), corners, found)
```

```python
display_img(flat_chess_bgr_cpy)

#corners are draw by connecting zigzag lines
```

```python
dots_grid = cv2.imread('Data/dot_grid.png')

display_img(dots_grid)
```

```python
#other grid pattern - like dot grid
#CALIB_CB_SYMMETRIC_GRID- default way of searching

found_dots, corners_dots = cv2.findCirclesGrid(dots_grid, (10,10), cv2.CALIB_CB_SYMMETRIC_GRID)

if found_dots:
    print(corners_dots)
else:
    pass
```

```python
corners_dots.shape
```

```python
dots_grid_cpy = dots_grid.copy()

cv2.drawChessboardCorners(dots_grid_cpy, (10,10), corners_dots, found_dots)

display_img(dots_grid_cpy)
```
